```python
#11)
import pandas as pd

data = {
    2014: [100.5, 150.8, 200.9, 30000, 40000],
    2015: [12000, 18000, 22000, 30000, 45000],
    2016: [20000, 50000, 70000, 100000, 125000],
    2017: [50000, 60000, 70000, 80000, 90000]
}
sales = pd.DataFrame(data, index=['Madhu', 'Kusum', 'Kinshuk', 'Ankit', 'Shruti'])

# a) Append the DataFrame Sales2 to the DataFrame Sales.
sales2_data = {
    2014: [135.6, 189.2],
    2015: [15000, 19500],
    2016: [21000, 52000],
    2017: [55000, 85000]
}
sales2 = pd.DataFrame(sales2_data, index=['Ravi', 'Divya'])
sales = pd.concat([sales, sales2])
print("After appending Sales2:\n", sales)

# b) Change the DataFrame Sales such that it becomes its transpose.
sales = sales.T
print("Transposed Sales:\n", sales)

# c) Display the sales made by all sales persons in the year 2017.
print("Sales in 2017:\n", sales.loc[2017])

# d) Display the sales made by Madhu and Ankit in the year 2017 and 2018.
print("Sales by Madhu and Ankit in 2017 and 2018:\n", sales.loc[[2017], ['Madhu', 'Ankit']])

# e) Display the sales made by Shruti 2016.
print("Sales by Shruti in 2016:\n", sales.loc[2016, 'Shruti'])
```

```python
#8)Write a Python program to create a plot to get a general Statistics of Iris data.
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('Iris.csv')
df.describe().plot(kind='bar')
plt.title('General Statistics of IRIS Data')
plt.tight_layout()
plt.show()

```

```python
#9)Write a Python program to create a Bar plot to get the frequency of the three species of the Iris data.
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv('Iris.csv')
df['Species'].value_counts().plot(kind='bar')
plt.title("Frequency of Each Iris Species")
plt.xlabel("Species")
plt.ylabel("Count")
plt.show()
```

```python
#10)Write a Python program to create a Pie plot to get the frequency of the three species of the Iris data.
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv('Iris.csv')
df['Species'].value_counts().plot(kind='pie', autopct='%1.1f%%')
plt.title("Species Distribution in Iris Data")
plt.ylabel("")
plt.show()
```

```python
#11)Write a Python program to create a graph to find relationship between the sepal length and width.
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv('Iris.csv')
plt.scatter(df['SepalLengthCm'], df['SepalWidthCm'])
plt.xlabel("Sepal Length")
plt.ylabel("Sepal Width")
plt.title("Sepal Length vs Sepal Width")
plt.show()
```

```python
#12)Write a Python program to create a graph to find relationship between the petal length and width.
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv('Iris.csv')
plt.scatter(df['PetalLengthCm'], df['PetalWidthCm'])
plt.xlabel("Petal Length")
plt.ylabel("Petal Width")
plt.title("Petal Length vs Petal Width")
plt.show()
```

```python
#13)Write a Python program to create a graph to see how the length and width of SepalLength, SepalWidth, PetalLength, PetalWidth are distributed.
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
df = pd.read_csv('Iris.csv')
sns.pairplot(df[['SepalLengthCm', 'SepalWidthCm', 'PetalLengthCm', 'PetalWidthCm']])
plt.show()
```

```python
#14)Write a Python program to create a joinplot to describe individual distributions on the same plot between Sepal length and Sepal width.
import pandas as pd
import seaborn as sns
df = pd.read_csv('Iris.csv')
sns.jointplot(x='SepalLengthCm', y='SepalWidthCm', data=df)
```

```python
#15)Write a Python program to create a joinplot using "hexbin" to describe individual distributions on the same plot between Sepal length and Sepal width.
import pandas as pd
import seaborn as sns
df = pd.read_csv('Iris.csv')
sns.jointplot(x='SepalLengthCm', y='SepalWidthCm', data=df, kind='hex')
```
